#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
cuadimg = cv2.imread('../DATA/dog_backpack.png')
cuadimg = cv2.cvtColor(cuadimg,cv2.COLOR_BGR2RGB)
```

```python
vertices = np.array([[440,420],[250,700],[625,700]],np.int32,)
pts = vertices.reshape((-1,1,2))
cv2.polylines(cuadimg,[pts],isClosed=True,color=(0,0,255),thickness=20)
plt.imshow(cuadimg)
```

```python
cuadimg = cv2.imread('../DATA/dog_backpack.png')
#cuadimg = cv2.cvtColor(cuadimg,cv2.COLOR_BGR2RGB)
plt.imshow(cuadimg)
#cv2.namedWindow(winname='new')
#cv2.imshow('new',cuadimg)
```

```python
cuadimg = cv2.resize(cuadimg,(311,467))
```

```python
import cv2
import numpy as np

##############
## FUNCTION ##
##############
def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(cuadimg,(x,y),40,(0,0,255),2)

cv2.namedWindow(winname='my_drawing')

cv2.setMouseCallback('my_drawing',draw_circle)


####################
## SHOWING IMAGES ##
####################
cuadimg = cv2.imread('../DATA/dog_backpack.png')
#cuadimg = cv2.cvtColor(cuadimg,cv2.COLOR_BGR2RGB)
cuadimg = cv2.resize(cuadimg,(311,467))

while True:
    cv2.imshow('my_drawing',cuadimg)  # Name of actual window
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```
