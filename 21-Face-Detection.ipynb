# Face detection

## Viola-Jones Algorithm with Haar Cascades

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nadia = cv2.imread('DATA/Nadia_Murad.jpg',0)
denis = cv2.imread('DATA/Denis_Mukwege.jpg',0)
solvay = cv2.imread('DATA/solvay_conference.jpg',0)
```

```python
plt.imshow(solvay, 'gray')
```

### Import pre-trained classifier

```python
face_cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img = img.copy()
    
    face_rects = face_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    
    return face_img
```

```python
result = detect_face(denis)
plt.imshow(result,'gray')
```

```python
result = detect_face(nadia)
plt.imshow(result,'gray')
```

```python
result = detect_face(solvay)
plt.figure(figsize=(14,7))
plt.imshow(result,'gray')
```

```python
def adj_detect_face(img):
    face_img = img.copy()
    
    face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.2, minNeighbors=5)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    
    return face_img
```

```python
result = adj_detect_face(solvay)
plt.figure(figsize=(14,7))
plt.imshow(result,'gray')
```

```python
eye_cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_eye.xml')
```

```python
def detect_eyes(img):
    face_img = img.copy()
    
    eyes_rects = eye_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)
    
    for (x,y,w,h) in eyes_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    
    return face_img
```

```python
result = detect_eyes(nadia)
plt.imshow(result,'gray')
```

# With webcam

```python
face_cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img = img.copy()
    
    face_rects = face_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(0,255,0),10)
    
    return face_img
```

```python
cap = cv2.VideoCapture(0)

while True:
    
    ret, frame = cap.read(0)
    
    frame = detect_face(frame)
    
    cv2.imshow('Video Face Detection', frame)
    
    k = cv2.waitKey(1)
    
    if k == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
