# Keras basics

```python
import numpy as np
from numpy import genfromtxt
```

```python
data = genfromtxt('DATA/bank_note_data.txt',delimiter=',')
```

```python
data
```

```python
labels = data[:,4]
labels
```

```python
features = data[:,0:4]
features
```

```python
X = features
y = labels
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

# Standardizing the Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_X_train = scaler.transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

# Building the Network with Keras

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Creates model
model = Sequential()
# 8 Neurons, expects input of 4 features. 
# Play around with the number of neurons!!
model.add(Dense(4, input_dim=4, activation='relu'))
# Add another Densely Connected layer (every neuron connected to every neuron in the next layer)
model.add(Dense(8, activation='relu'))
# Last layer simple sigmoid function to output 0 or 1 (our label)
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(scaled_X_train, y_train, epochs=50, verbose=2)
```

```python
model.predict_classes(scaled_X_test)
```

```python
model.metrics_names
```

# Performance

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
preds = model.predict_classes(scaled_X_test)
```

```python
confusion_matrix(y_test, preds)
```

```python
print(classification_report(y_test, preds))
```

# Save Model

```python
model.save('first_model.h5')
```
