# Blending:

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('DATA/dog_backpack.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

img2 = cv2.imread('DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1) 
```

```python
plt.imshow(img2)
```

```python
img1.shape
```

```python
img2.shape
```

```python
# resize both to same size
```

```python
img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

# $\quad$ new_pixel = $\alpha$ $\times$ pixel1 + $\beta$ $\times$ pixel2 + $Y$

```python
blended = cv2.addWeighted(src1=img1, alpha=0.8, src2=img2 , beta=0.1, gamma=0)
plt.imshow(blended)
```

## Overlay small on top of large image (no blend)

```python
# Numpy reassignment
img1 = cv2.imread('DATA/dog_backpack.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

img2 = cv2.imread('DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2,(600,600))
plt.imshow(img2)
```

```python
large_img = img1
small_img = img2
```

```python
x_offset = 0
y_offset = 0

x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]

large_img[y_offset:y_end, x_offset:x_end] = small_img

plt.imshow(large_img)
```
